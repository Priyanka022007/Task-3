```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import file for task
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

# Categorical Variable Visualization

```python
#Employee count per department
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title("Employee count per Department")
```

 Finance employees are more than Marketting and Engineering

```python
#To Count Gender i.e male or female

sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title("Employee count per Gender")
```

 Female Employees are more than male employees

```python
#Employee location Distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title("Employee count per Location")
plt.xticks(rotation=45)
plt.show()
```

# Numerical Variable Visualization

```python
#Age Distribution
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title("Employees count as per Age")
plt.show()
```

More picks of age in 45 to 50 age followed by freshers in 25 to 30 age

```python
#Years of Experiance Distribution
sns.histplot(ds['Years_of_Experience'],bins=10,kde=True)
plt.title("Employees count as per Years_of_Experience")
plt.show()
```

```python
#Monthly salary Distribution
sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title("Employee count per Monthly_Salary")
```

```python
#Average Salary by Department
sns.barplot(data=ds, x="Department", y="Monthly_Salary", estimator=np.mean, palette='viridis')
plt.title("Average Salary by Department")
plt.show()
```

```python
#Perform Score by Department
sns.boxplot(data=ds, x="Department", y="Performance_Score", palette='pastel')
plt.title("Performance score by Department")
plt.show()
```

```python
#Salary Distribution by Gender
sns.violinplot(data=ds, x="Gender", y="Monthly_Salary", palette='muted')
plt.title("Salary by Gender")
plt.show()
```

Male has More salary followed by female

```python
#Correlation matrix
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot = True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Salary vs Experience Scatter plot
sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
plt.title("Salary vs Experience by Department")
plt.show()
```

 as more experience finance employees have more salary
